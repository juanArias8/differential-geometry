```python
%matplotlib inline
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

@np.vectorize
def evaluateF1x(x_value):
    return f1x.subs(x, x_value)

@np.vectorize
def evaluateF2x(x_value):
    return f2x.subs(x, x_value)

@np.vectorize
def evaluateF3x(x_value):
    return f3x.subs(x, x_value)

f1x = sym.sympify(raw_input('Ingrese una función en términos de x:'))
f2x = sym.sympify(raw_input('Ingrese una función en términos de x:'))
f3x = sym.sympify(raw_input('Ingrese una función en términos de x:'))
aRango = input("Ingrese el valor a del rango");
bRango = input("Ingrese el valor b del rango");
x = sym.Symbol("x")

vector = np.linspace(aRango, bRango)
rango = np.linspace(aRango, bRango)

f1xEvaluado = evaluateF1x(vector)
f2xEvaluado = evaluateF2x(vector)
f3xEvaluado = evaluateF3x(vector)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(f1xEvaluado, f2xEvaluado, f3xEvaluado, rstride=1, cstride=1, cmap='hot')
plt.show()
```

```python

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as pl

fig = pl.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=pl.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=pl.cm.hot)
ax.set_zlim(-2, 2)

pl.show()
```
